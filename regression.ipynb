```python
import numpy as np
import pandas
data = pandas.read_csv('data-logistic.csv', header=None)
print(data)
target = data[0]
pre_1 = np.array(data.loc[:, 1:1])
pre_2 = np.array(data.loc[:, 2:2])
signs_1 = []
signs_2 = []
for i in range(205):
    signs_1.append(pre_1[i][0])
    signs_2.append(pre_2[i][0])
print(signs_1)
print(signs_2)
```

```python
def regression(w1, w2, C):
    f1 = []
    f2 = []
    for i in range(205):
        f1.append(target[i]*signs_1[i] * (1-1/(1+np.exp((-target[i])*(w1*signs_1[i]+w2*signs_2[i])))))
        f2.append(target[i]*signs_2[i]*(1-1/(1+np.exp((-target[i])*(w1*signs_1[i]+w2*signs_2[i])))))
    w1 = (1-0.1*C) * w1 + 0.1/205 * np.sum(f1)
    w2 = (1-0.1*C) * w2 + 0.1/205 * np.sum(f2)
    return [w1, w2]
```

```python
w1 = 0.0
w2 = 0.0
C = 0
c = regression(w1, w2, C)
while (c[0] - w1)**2 + (c[1] - w2)**2 > 1e-10:
    w1 = c[0]
    w2 = c[1]
    c = regression(w1, w2, C)
w1 = c[0]
w2 = c[1]
print(w1, w2)

w3 = 0.0
w4 = 0.0
C = 10
c = regression(w3, w4, C)


while (c[0] - w3)**2 + (c[1] - w4)**2 > 1e-10:
    w3 = c[0]
    w4 = c[1]
    c = regression(w3, w4, C)
w3 = c[0]
w4 = c[1]
print(w3, w4)
```

```python
results = []
results_reg = []
for i in range (205):
    results.append(1 / (1 + np.exp(-w1 * signs_1[i] - w2 * signs_2[i])))
    results_reg.append(1 / (1 + np.exp(-w3 * signs_1[i] - w4 * signs_2[i])))
print(results)
print(results_reg)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(target, results))
print(roc_auc_score(target, results_reg))
```

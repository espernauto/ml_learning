```python
import numpy as np

import pandas

data = pandas.read_csv('close_prices.csv')
print(data)
needed = data.loc[:, 'AXP':'XOM']
print(needed)
```

```python
from sklearn.decomposition import PCA
```

```python
sum = 0
clf = PCA(n_components = 30)
clf.fit(needed)
    #if i != 0:
    #    sum = sum + np.array(clf.explained_variance_ratio_)[i]
    #print(clf.explained_variance_ratio_)
    #print(sum)
i = 0
print(clf.explained_variance_ratio_)
while sum < 0.9:
    print(clf.explained_variance_ratio_[i])
    sum = sum + clf.explained_variance_ratio_[i]
    i+=1
print(sum)
print(i)
```

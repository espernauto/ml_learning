```python
from sklearn.datasets import load_boston
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
```

```python
boston = load_boston()
print(boston.data)
print(boston.target)
```

```python
from sklearn.preprocessing import scale
data_scaled = scale(boston.data)
print(data_scaled)
```

```python
param = np.linspace(1, 10, 200)
print(param)
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
mean = []
X = np.array(data_scaled)
y = boston.target
print(X)
kf = KFold(shuffle = True, random_state = 42)
for i in param:
    dtc = KNeighborsRegressor(n_neighbors = 5, weights = 'distance', p = i)
    r = cross_val_score(dtc, X, y, cv=kf, scoring='neg_mean_squared_error')
    mean.append(r.mean())
print(mean)
result = mean.index(max(mean)) + 1
f = open('results.txt', 'w')
f.write(str(result))
```

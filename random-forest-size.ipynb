```python
import numpy as np

import pandas

data = pandas.read_csv('abalone.csv')
print(data)
```

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
```

```python
print(data)
```

```python
signs = data.loc[:, 'Sex':'ShellWeight']
target = data['Rings']
print(signs)
print(target)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import r2_score
```

```python
kf = KFold(random_state = 1, shuffle = True)
X = np.array(signs)
y = np.array(target)
for i in range(1, 50):
    clf = RandomForestRegressor(n_estimators = i, random_state = 1)
    clf.fit(X, y)
    r = cross_val_score(clf, X, y, cv=kf, scoring='r2')
    print(r.mean(), i) #можно глазами увидеть, можно было бы собрать в массив и вывести нужный. но данные маленькие, зачем
```
